# 机器学习核心概念笔记

## 一、监督学习（Supervised Learning）

**核心思想**：从"带标签"的数据中学习输入→输出的映射关系。

- **训练数据**：每条样本都有明确的标签/答案，形如 `(x, y)`
- **目标**：学习函数 `f(x) → y`，使模型能对新数据做出正确预测
- **典型任务**：
  - **分类**：判断邮件是否为垃圾邮件、图片中是猫还是狗
  - **回归**：预测房价、预测气温
- **常见算法**：线性回归、逻辑回归、SVM、决策树、随机森林、神经网络

**类比**：老师给学生批改过的试卷（带正确答案），学生通过对答案来学习。

## 二、无监督学习（Unsupervised Learning）

**核心思想**：从"无标签"的数据中发现隐藏的结构和模式。

- **训练数据**：只有输入 `x`，没有标签 `y`
- **目标**：发现数据内部的分布、分组、关联等规律
- **典型任务**：
  - **聚类**：将客户按消费行为自动分群
  - **降维**：将高维数据压缩到低维（如 PCA）
  - **异常检测**：发现不符合常规模式的数据点
- **常见算法**：K-Means、DBSCAN、层次聚类、PCA、自编码器

**类比**：没有老师，学生自己把一堆试卷按相似度分类整理。

## 三、监督 vs 无监督 对比

| 维度 | 监督学习 | 无监督学习 |
|------|---------|-----------|
| 数据 | 有标签 `(x, y)` | 无标签，只有 `x` |
| 目标 | 预测/分类 | 发现结构/模式 |
| 评估 | 容易（对比预测值与真实标签） | 较难（缺乏客观标准） |
| 成本 | 标注数据昂贵 | 数据获取成本低 |
| 典型场景 | 垃圾邮件过滤、医学诊断 | 用户分群、数据可视化 |

## 四、机器学习任务类型总览

| 任务类型 | 核心特点 | 输出形式 | 典型例子 |
|---------|---------|---------|---------|
| 分类 | 判断属于哪一类 | 离散类别 / 标签 | 垃圾邮件识别、猫狗分类、情感正负 |
| 回归 | 预测具体数值 | 连续实数 | 房价预测、气温预测、销售额预测 |
| 聚类 | 无标签自动分组 | 簇 / 群组 | 用户分群、新闻聚类、异常样本识别 |
| 目标检测 | 识别物体并定位 | 类别 + 位置框 | 图片中人车检测、病灶定位 |
| 语义分割 | 像素级区域划分 | 像素类别标签 | 自动驾驶路面分割、医学影像分割 |
| 时序预测 | 按时间预测未来 | 连续 / 离散时序值 | 用电量预测、股价走势、交通流量 |
| 生成任务 | 创造新数据 | 全新文本 / 图像 / 音频 | AI 绘画、文本生成、语音合成 |
| 推荐系统 | 匹配用户偏好 | 推荐列表 / 评分 | 电商猜你喜欢、短视频推荐 |
| 强化学习 | 试错学习最优策略 | 动作 / 决策序列 | AlphaGo、机器人控制、游戏 AI |
| 降维 | 简化数据保留信息 | 低维特征向量 | 数据可视化、高维特征压缩 |

# 机器学习基础认知篇

## 一、科研和机器学习到底在干嘛？

**一句话：让计算机从数据里自己学会"规律"，然后用这个规律去预测新东西。**

### 传统编程 vs 机器学习

| 方式 | 逻辑 | 例子 |
|------|------|------|
| 传统编程 | 人写规则 + 数据 → 结果 | `if temperature > 37: print("发烧")` |
| 机器学习 | 数据 + 结果 → 机器自己找规则 | 给一堆体温+是否发烧的数据，机器自己学会判断 |

### "模型"是什么？

模型 = 一个**带未知参数的数学公式**。

最简单的例子：`y = wx + b`
- `x`：输入（如房子面积）
- `y`：输出（如房价）
- `w` 和 `b`：参数，一开始不知道，靠训练确定

**训练 = 用大量数据，让计算机不断调整 w 和 b，使预测的 y 尽可能接近真实的 y。**

### 科研/论文在干嘛？

科研三步走：
1. **发现问题**：现有方法效果不够好（预测不准、速度太慢）
2. **提出新方法**：改进模型结构、换训练方式、用新的数据处理技巧
3. **实验证明更好**：跑数据，对比指标，列表格

读论文核心看三件事：
- **解决什么问题？**（Introduction）
- **怎么做的？**（Method — 架构图 + 公式）
- **效果比别人好多少？**（Experiments — 表格数字）

## 二、机器学习三大环节

做机器学习就像做菜，分三个环节：

| 环节 | 类比 | 机器学习 |
|------|------|---------|
| 数据处理 | 洗菜、切菜、备料 | 把原始数据整理成模型能吃的格式 |
| 模型结构 | 用什么锅、什么灶 | 用什么数学公式/网络架构来学习 |
| 训练方式 | 火候怎么调、炒多久 | 怎么优化参数、训练多少轮 |

### 2.1 数据处理（Data Processing）

原始数据往往是"脏"的，不能直接用，常见操作：

| 操作 | 做什么 | 例子 |
|------|--------|------|
| 清洗 | 删缺失值、去错误数据 | 身高 999cm 明显是错的，删掉 |
| 标准化/归一化 | 缩放到统一范围 | 身高 170cm 和体重 60kg 差距太大，都变成 0~1 |
| 特征提取 | 从原始数据挑有用信息 | 从文字提取关键词，从图片提取边缘 |
| 数据增强 | 人为造更多数据 | 把图片翻转、旋转、加噪声 |
| 划分数据集 | 分成三份 | 训练集（学习）、验证集（调参）、测试集（考试） |

### 2.2 模型结构（Model Architecture）

选择用什么"公式"来拟合数据，从简单到复杂：

| 模型 | 一句话理解 |
|------|-----------|
| 线性模型 | `y = wx + b`，一条直线 |
| 决策树 | 像问答树："身高>170？→ 是 → 体重>70？→ 预测：男" |
| 神经网络 | 多层节点互相连接，层越多越"深"= 深度学习 |
| CNN | 专门处理图片，自动识别边缘、纹理、形状 |
| Transformer | 当前最火，GPT/BERT 的核心，靠"注意力机制"学会重点看哪里 |

**论文说"提出新模型结构" = 设计了一种新的公式/网络连接方式，效果比以前的好。**

### 2.3 训练方式（Training Strategy）

模型选好了，还要决定怎么训练：

| 概念 | 含义 | 直觉理解 |
|------|------|---------|
| 学习率 | 每次调参数的步子多大 | 太大跳过最优解，太小训练太慢 |
| 批大小（Batch Size） | 每次拿多少条数据算一次 | 通常 32 或 64，兼顾速度和稳定 |
| 训练轮数（Epoch） | 整个数据集过几遍 | 过少学不好，过多死记硬背（过拟合） |
| 优化器（Optimizer） | 梯度下降的具体算法 | Adam 最常用 |
| 预训练+微调 | 先用大数据训通用模型，再用小数据调 | GPT 就是这么训练的 |

## 三、损失函数（Loss Function）详解

### 核心问题：模型预测得准不准？

损失函数 = 一个**打分标准**，量化"预测值"和"真实值"之间的差距。

**损失越大 = 预测越烂，损失越小 = 预测越准。训练的目标就是让损失尽可能小。**

### 最直觉的例子：均方误差（MSE）

| 房子 | 真实价格 | 预测价格 | 差距 | 差距² |
|------|---------|---------|------|-------|
| A | 100万 | 90万 | -10 | 100 |
| B | 200万 | 210万 | +10 | 100 |
| C | 150万 | 150万 | 0 | 0 |

**MSE = (100 + 100 + 0) / 3 = 66.7**

为什么要**平方**？
1. 消除正负号（+10 和 -10 加起来变 0，看起来完美但实际不是）
2. 放大大误差——差 10 万比差 1 万严重得多，平方后惩罚更重

### 不同任务用不同的损失函数

| 任务 | 损失函数 | 直觉理解 |
|------|---------|---------|
| 回归（预测数值） | MSE（均方误差） | 预测值和真实值差多远 |
| 二分类（是/否） | BCE（二元交叉熵） | 预测的概率和真实标签差多远 |
| 多分类（猫/狗/鸟） | CrossEntropy（交叉熵） | 预测的概率分布和真实分布差多远 |

**记住一句话：损失函数 = 衡量"错得多离谱"的数字。**

## 四、梯度下降（Gradient Descent）详解

### 问题：参数有无数种组合，怎么找到最好的？

**类比：蒙着眼站在山上，要走到最低点（谷底）**

- 感受脚下坡度（**梯度**）→ 往低处走
- 坡度越陡 → 步子越大
- 脚下平了 → 到谷底了，停

对应关系：
- **山的高度** = 损失函数的值（越低越好）
- **你的位置** = 当前参数值
- **坡度** = 梯度（损失函数对参数的偏导数）
- **步子大小** = 学习率

### 核心公式（就这一行）

```
新参数 = 旧参数 - 学习率 × 梯度
```

反复执行这一行，参数就会慢慢"走"到让损失最小的位置。

### 学习率的影响

```
学习率太大：  ↗ ↘ ↗ ↘   在谷底两边跳来跳去，永远到不了
学习率太小：  →→→→→→→   慢慢挪，训练一万年
学习率合适：  →  →  → ·  几步就稳稳到达谷底
```

### 梯度下降的变种

| 名称 | 做法 | 特点 |
|------|------|------|
| BGD（批量梯度下降） | 用全部数据算一次梯度 | 准确但巨慢 |
| SGD（随机梯度下降） | 每次只用 1 条数据 | 快但不稳定，像醉汉走路 |
| Mini-batch SGD | 每次用一小批（如 32 条） | **最常用**，兼顾速度和稳定 |
| Adam | 自动调整每个参数的学习率 | **论文默认选择** |

## 五、训练循环全流程

把所有环节串起来，这就是 ML 代码的核心骨架：

```
1. 准备数据（数据处理）
2. 选一个模型结构（比如神经网络）
3. 初始化参数（随机值）
4. 拿一批数据，模型做预测         → 前向传播 (forward)
5. 用损失函数算"预测有多烂"       → 计算 loss
6. 算梯度：每个参数该往哪个方向调  → 反向传播 (backward)
7. 用梯度下降更新参数             → optimizer.step()
8. 重复 4-7，直到损失足够小
```

对应的 PyTorch 代码：

```python
for epoch in range(100):               # 训练 100 轮
    for batch in dataloader:           # 每次取一小批数据
        pred = model(batch.x)          # 4. 前向传播
        loss = loss_fn(pred, batch.y)  # 5. 算损失
        loss.backward()                # 6. 反向传播，算梯度
        optimizer.step()               # 7. 更新参数
        optimizer.zero_grad()          # 清空梯度，准备下一轮
```

**所有 ML 代码都是这个结构，没有例外。**

## 六、学习 ML 的正确姿势与习惯

### 核心原则：看懂不算会，跑通不算会，改着玩才算会

每学一个算法/概念，都走三步循环：

```
看懂（30分钟）→ 手写代码跑通（1-2小时）→ 改参数玩（30分钟-1小时）
```

70% 的时间花在"动手改"上，而不是"看视频"上。

---

### 习惯一：永远先跑通，再理解

不要上来就研究公式推导。先把代码跑通看到结果，再倒回去理解每一行在干嘛。

```
❌ 看公式 → 推导 → 写代码 → 跑（卡在第一步就放弃了）
✅ 跑代码 → 看结果 → 改参数 → 倒回去看公式为什么是这样
```

### 习惯二：用 Jupyter Notebook，一个 cell 一个 cell 跑

不要写完一整个文件再运行。一小块一小块执行，每跑一块都能看到中间结果：

```
Cell 1: 加载数据 → 打印看看数据长什么样
Cell 2: 画个图 → 直觉感受数据分布
Cell 3: 定义模型
Cell 4: 训练 → 看 loss 在下降
Cell 5: 预测 → 和真实值对比
```

### 习惯三：每学一个算法，画一张手写流程图

不用好看，纸上随便画。能画出来 = 真的理解了，画不出来 = 还没懂。

```
线性回归流程图：
输入x → 乘以w加b → 预测值 → 和真实值比较 → 算loss → 算梯度 → 更新w,b → 循环
```

### 习惯四：建立"对比学习"的习惯

每学一个新东西，立刻和已经会的东西对比：

```
学逻辑回归时 → 和线性回归对比：多了什么？（Sigmoid + 交叉熵）
学决策树时   → 和逻辑回归对比：思路完全不同（规则 vs 概率）
学 CNN 时    → 和 MLP 对比：多了什么？（卷积层 + 池化层）
学 LSTM 时   → 和普通 RNN 对比：解决了什么问题？（长期记忆）
```

### 习惯五：遇到不懂的，先"容忍"再"回填"

不要卡住。80% 的情况下跳过不影响理解核心思想。

```
看论文遇到 "KL 散度" → 标记一下，跳过，继续往下看
后面理解了大意后     → 再回来查 KL 散度到底是什么
```

带着上下文去查，反而学得更快。

### 习惯六：记笔记只记"自己的理解"，不抄定义

```
❌ 梯度下降：一种通过计算损失函数梯度来迭代更新参数的优化算法
✅ 梯度下降：蒙眼下山，摸坡度走，学习率就是步子大小，调大了会跳来跳去
```

第一种是抄书，以后看还是不懂。第二种是自己消化过的，一看就想起来。

## 七、深度学习代码在干什么？

**所有 DL 代码不管多长，永远只在干五件事：**

| 阶段 | 在干嘛 | 关键词 | 代码量占比 |
|------|--------|--------|-----------|
| 准备数据 | 把原始数据变成模型能吃的数字矩阵 | `Dataset`, `DataLoader`, `transform` | 30-40% |
| 定义模型 | 搭积木：几层、每层多少节点、什么激活函数 | `nn.Module`, `forward`, `nn.Linear`, `nn.Conv2d` | 20-30% |
| 定损失和优化器 | 怎么打分、怎么更新参数 | `Loss`, `optim`, `Adam`, `lr=` | 就两三行 |
| 训练循环 | 核心五步，套路固定 | `backward()`, `step()`, `zero_grad()` | 核心 |
| 评估/预测 | 拿测试数据看效果 | `eval()`, `no_grad()`, `accuracy` | 10-20% |

### 训练循环永远是这五步

```python
pred = model(batch_x)             # ① 前向传播：喂数据，得到预测
loss = loss_fn(pred, batch_y)      # ② 算损失：预测和真实答案差多少
optimizer.zero_grad()              # ③ 清空上一轮的梯度
loss.backward()                    # ④ 反向传播：算每个参数的梯度
optimizer.step()                   # ⑤ 更新参数：沿梯度方向走一步
```

### 看起来吓人的代码实际在干嘛？

| 看起来吓人的部分 | 实际在干嘛 |
|----------------|-----------|
| `argparse` 一大堆参数 | 命令行配置，让你能改参数 |
| 各种 `if` 判断 | 处理不同情况（GPU/CPU、恢复训练等） |
| `logging`、`wandb`、`tensorboard` | 记录日志、画曲线图 |
| `save`/`load` checkpoint | 保存/加载训练好的模型 |
| 各种 `utils.py` | 工具函数（画图、计算指标） |
| `config.yaml` | 配置文件（超参数统一管理） |

**这些全是"脚手架"，不是核心。核心永远是那五步训练循环。**

### 看别人代码的方法：搜关键词定位

| 搜这个词 | 就找到了 |
|---------|---------|
| `DataLoader` | 数据准备部分 |
| `def forward` | 模型定义部分 |
| `Loss` | 损失函数 |
| `optim` | 优化器 |
| `backward()` | 训练循环核心 |
| `eval()` | 评估部分 |

## 八、科研的创新到底在哪？

### 核心认知：创新在"想法"上，不在代码上

代码是工具，所有人都用 PyTorch，训练循环都长一样。区别在于**你在哪个环节做了不一样的事**。

### 五种创新方式

| 环节 | 创新方式 | 真实例子 |
|------|---------|---------|
| 数据层面 | 构建新数据集或新的数据处理方式 | 发现中文电商有"反话"差评，构建反讽情感数据集 |
| 模型结构 | 在现有模型基础上加/换模块 | 在 ResNet 基础上插一个 Attention 模块，效果更好 |
| 训练方式 | 换一种训练策略 | 先用无标注数据自监督预训练，再用少量标注数据微调 |
| 问题定义 | 用 ML 解决一个别人没想到的问题 | 用 ML 预测文件协作中哪些文件可能冲突，提前预警 |
| 跨领域迁移 | 把 A 领域的方法用到 B 领域 | Transformer 从翻译迁移到图像 → ViT，发了顶会 |

**跨领域迁移是最常见的论文创新方式。** 代码几乎没改，但"想到把它用在这里"就是创新。

### 创新的实际流程

```
1. 大量读论文 → 知道别人做到什么程度、还有什么问题没解决
2. 发现一个痛点 → "这个方法在 XX 场景下效果不好"
3. 想到一个改进 → "如果在这里加/换一个 XX 会不会更好？"
4. 用现有代码改几十行实现你的想法
5. 跑实验对比 → 证明你的方法确实更好
6. 写论文
```

**第 1-3 步是创新，第 4 步只是实现。你的价值在于"想到"，不在于"写出来"。**

### "改代码"改的是逻辑，不是参数

原始论文的模型：
```python
class OriginalModel(nn.Module):
    def forward(self, x):
        x = F.relu(self.cnn(x))       # 卷积提特征
        x = self.fc(x)                # 全连接做分类
        return x
```

你的改进——加一个注意力模块：
```python
class ImprovedModel(nn.Module):
    def forward(self, x):
        x = F.relu(self.cnn(x))       # 卷积提特征
        x = self.attention(x)          # ← 你的创新点：加了注意力
        x = self.fc(x)                # 全连接做分类
        return x
```

代码只多了一行，但**这一行就是你整篇论文的核心贡献。**

### 一句话总结

```
代码是别人的   → 正常，所有人都站在巨人肩膀上
参数是调的     → 正常，这叫"实验"
创新在想法上   → 你发现了什么问题？提出了什么改进？为什么你的方法更好？
```

**科研不是比谁代码写得多，是比谁想法好。代码只是验证想法的工具。**

## 九、科研 + AI 辅助的最佳工作流

### 总览：六阶段循环

```
① 选方向 → ② 读论文 → ③ 数学按需补 → ④ 写代码 → ⑤ 实验对比 → ⑥ 沉淀输出
     ↑                                                              │
     └──────────────── 发现新问题/新想法 ←──────────────────────────┘
```

---

### 阶段一：选方向——"我要解决什么问题"

不要上来就看论文，先搞清楚做什么。

```
1. 确定研究领域关键词（如：图像分类、文本生成、推荐系统）
2. 问 AI："XX 领域目前有哪些主流方法？优缺点？还有什么未解决的问题？"
3. AI 给你全景地图，从中选一个方向
```

**AI 角色：领域导游。** 5 分钟了解别人几个月才摸清的全貌。

---

### 阶段二：读论文——"别人做了什么"

找 3-5 篇核心论文，每篇用**三遍阅读法**：

| 遍数 | 时间 | 看什么 | 目标 |
|------|------|--------|------|
| 第一遍 | 5 分钟 | Abstract + 架构图 + 实验表格 | 知道干了什么、效果如何 |
| 第二遍 | 30 分钟 | Method 部分，不懂的公式截图问 AI | 理解具体怎么做的 |
| 第三遍 | 按需 | 找代码跑一遍 | 从代码层面理解实现 |

每读一篇，用自己的话写 3-5 句总结：
- 解决什么问题？
- 核心方法是什么？
- 效果怎么样？
- **还有什么局限？** ← 最重要，这就是你未来的创新点来源

**AI 用法：**

| 你遇到的问题 | 怎么问 AI |
|-------------|----------|
| 公式看不懂 | 截图："用大白话解释这个公式在干什么" |
| 英文读不动 | 贴段落："翻译并解释核心意思" |
| 不知道某个方法 | "XX 方法的核心思想？和 YY 有什么区别？" |
| 不知道该读哪篇 | "XX 领域最有影响力的 5 篇论文是哪些？" |

**关键习惯：读完每篇论文，花 2 分钟记一条"局限"。积累 5-10 条，创新点自然浮现。**

---

### 阶段三：数学按需补——"遇到再查，不系统学"

**永远不要单独花时间"学数学"。**

```
1. 读论文/写代码 → 遇到不懂的数学
2. 先标记，不停下来
3. 读完后集中问 AI：
   "交叉熵的直觉理解？为什么分类用它不用 MSE？"
   "softmax 到底算什么？给我一个具体数字例子"
4. AI 给直觉解释 + 数字例子 → 够用了
5. 反复出现的概念才值得深入推导
```

```
❌ 打开高数课本从第一章学
✅ 遇到 → 问 AI 要直觉解释和数字例子 → 理解 → 继续
```

你需要的不是"会推导"，而是"看到公式知道它在干嘛"。

---

### 阶段四：写代码——"跑通别人的，改成自己的"

```
1. 找论文对应的开源代码（GitHub / Papers With Code）
2. 先原封不动跑通
3. 跑通后读代码，对照论文理解每部分
4. 开始改：
   - 先改数据 → 换成自己的数据集
   - 再改模型 → 加你的改进模块
   - 最后调参 → 找最好的超参数
5. 每改一次跑一次，对比结果
```

**AI 用法：**

| 阶段 | 怎么用 AI |
|------|----------|
| 环境搭建 | "Windows + PyTorch + CUDA 12.x，写安装命令" |
| 报错调试 | 直接贴报错信息 |
| 读代码 | "这段代码在干什么？逐行解释" |
| 改代码 | "在第二层后加 Self-Attention，帮我写" |
| 调参 | "这个模型用什么学习率和 batch size 合适？" |
| 画图 | "帮我写代码画 loss 曲线和 accuracy 曲线" |

---

### 阶段五：实验对比——"证明你的方法更好"

```
1. 确定 Baseline（对比基准）：读过的那几篇论文的方法
2. 设计实验表格：
   | 方法 | 准确率 | F1 | 推理速度 |
   |------|--------|-----|---------|
   | 方法A（别人的） | 85.2% | 0.83 | 50ms |
   | 方法B（别人的） | 87.1% | 0.86 | 80ms |
   | 你的方法        | 89.3% | 0.88 | 60ms |
3. 做消融实验：去掉你加的模块，效果下降 → 证明模块有用
4. 效果不好？问 AI 分析原因 → 调整 → 重新跑（试错本身就是科研）
```

## 十一、人工智能完整知识地图

### 最高层：AI 的分支关系

```
人工智能（AI）
├── 机器学习（ML）—— 从数据中学规律
│   ├── 传统机器学习 —— 经典算法，小数据也能用
│   ├── 深度学习（DL）—— 神经网络，需要大数据
│   └── 强化学习（RL）—— 通过试错学最优策略
└── 非 ML 的 AI —— 专家系统/规则引擎（已过时）
```

**深度学习和强化学习都属于机器学习，不是独立的东西。**

### 11.1 传统机器学习——问题→算法映射

```
你有数据，想得到什么？
├── 预测数字 ────→ 回归：线性回归、岭回归、决策树回归、随机森林回归
├── 判断类别 ────→ 分类：逻辑回归、KNN、SVM、决策树、随机森林、XGBoost
├── 发现分组 ────→ 聚类：K-Means、DBSCAN、层次聚类
├── 简化数据 ────→ 降维：PCA、t-SNE、UMAP
└── 找异常 ──────→ 异常检测：Isolation Forest、One-Class SVM
```

**算法进化关系：**
```
线性回归 → +Sigmoid → 逻辑回归
决策树 → 多棵投票 → 随机森林 → +Boosting → XGBoost（表格数据之王）
```

**什么时候用什么：**

| 数据类型 | 数据量 | 推荐 |
|---------|--------|------|
| 表格（CSV） | 任意 | XGBoost / 随机森林 |
| 图片 | 大量 | CNN（深度学习） |
| 文本 | 大量 | Transformer（深度学习） |
| 少量数据 | <1000条 | 逻辑回归 / SVM / 随机森林 |

### 11.2 深度学习——神经网络进化树

```
感知机（1957）→ MLP 多层感知机（1986）
  │
  ├─→ CNN 卷积神经网络（图像）
  │    LeNet → AlexNet → VGG → ResNet → YOLO（检测）/ U-Net（分割）
  │
  ├─→ RNN 循环神经网络（序列）
  │    RNN → LSTM（解决长期记忆）→ GRU（LSTM简化版）
  │
  ├─→ Transformer（2017，当前王者）
  │    ├─ BERT（双向理解）→ 文本分类/问答
  │    ├─ GPT（单向生成）→ ChatGPT/写作/对话
  │    ├─ ViT（图像版Transformer）
  │    └─ CLIP（图文多模态）
  │
  ├─→ GAN 生成对抗网络（生成器 vs 判别器互相博弈）
  ├─→ VAE 变分自编码器（学分布→采样生成）
  └─→ Diffusion 扩散模型（加噪→去噪）→ Stable Diffusion / DALL-E
```

**关键进化关系：**
```
RNN记不住长文 → LSTM加门能记住 → 但太慢
    → Transformer全靠注意力，能并行 → 碾压RNN/LSTM
        → BERT（理解）+ GPT（生成）
```

**什么问题用什么网络：**

| 问题 | 用什么 |
|------|--------|
| 图片分类/检测 | CNN（ResNet / YOLO） |
| 文本理解 | BERT |
| 文本生成 | GPT |
| 图片生成 | Diffusion / GAN |
| 语音识别 | Transformer（Whisper） |
| 多模态（图+文） | CLIP / 多模态大模型 |
| 表格数据 | **不用DL，用XGBoost** |

### 11.3 强化学习——完全不同的思路

```
ML/DL：给数据和答案 → 学规律 → 预测
RL：   没有标准答案 → 自己尝试 → 根据奖惩调整策略
```

核心概念：Agent 观察 State → 做 Action → 得到 Reward → 进入新 State → 循环

```
强化学习
├── 基于价值：Q-Learning → DQN（神经网络版）
├── 基于策略：REINFORCE → Policy Gradient
├── 两者结合：A2C/A3C → PPO（最常用，RLHF就用它）
└── 应用：游戏AI、机器人、推荐系统、大模型对齐（RLHF）
```

### 11.4 技术栈——用什么工具

```
数据处理：NumPy（矩阵）、Pandas（表格）、Matplotlib（画图）
传统ML：  scikit-learn（一行代码调用经典算法）
深度学习：PyTorch（学术主流）、TensorFlow（工业界）
预训练模型：HuggingFace（ML界的GitHub，几十万个模型）、timm（图像模型库）
实验管理：wandb、TensorBoard、MLflow
```

### 11.5 全景决策树——拿到问题怎么选

```
现实问题 → 判断类型 → 选方法 → 选工具
├── 表格数据？ ──→ 传统ML ──→ scikit-learn + XGBoost
├── 图片？ ─────→ CNN ──────→ PyTorch + timm
├── 文本？ ─────→ Transformer → PyTorch + HuggingFace
├── 生成图片？ ─→ Diffusion ─→ Stable Diffusion
├── 生成文本？ ─→ GPT类 ────→ HuggingFace / API
├── 游戏/控制？ → RL ────────→ Gymnasium + Stable-Baselines3
└── 不确定？ ──→ 先试随机森林/XGBoost，不行再上DL
```

**所有方法共享的底层逻辑：数据 → 模型 → 损失函数 → 优化 → 评估**

## 十二、ML vs DL 的本质区别

**只有一个区别：谁来提取特征。**

```
传统 ML：人工提取特征 → 算法学规律
深度学习：算法自己提取特征 → 自己学规律
```

CNN 为例，每层自动学到的东西：
```
第1层：边缘（横线、竖线）
第2层：纹理（条纹、圆点）
第3层：部件（眼睛、耳朵）
第4层：整体（猫脸、狗脸）
```

**什么时候用什么：**
- 数据大 + 图片/文本/语音 → 深度学习
- 数据小 + 表格 → 传统 ML（XGBoost）
- 需要可解释性（医疗/金融）→ 决策树等传统 ML

## 十三、科研中哪些地方需要巧思

### 真正需要动脑的三个地方

**1. 问题定义（最值钱）**
- 不是"怎么提 1% 准确率"，而是"问题本身有没有定义错"
- 纯靠领域理解和观察，不需要代码能力

**2. 方法设计（组合的艺术）**
- 不是从零发明，而是有策略地组合现有模块
- 方法 A 擅长局部 + 方法 B 擅长全局 → 串联起来 → 一篇论文

**3. 实验分析（深挖为什么）**
- 在什么样本上提升最大？为什么？
- 什么情况下变差了？能不能解决？
- 消融实验：去掉模块 A 掉 3%，去掉 B 掉 0.5% → A 是核心贡献

### 不需要巧思的（交给 AI）

数据清洗、训练脚本、画图做表、环境配置、debug、论文排版

### 精力分配

```
问题定义/选题       ★★★★★  10%
读论文找 gap        ★★★★☆  20%
方法设计            ★★★★★  15%
数据准备            ★★★☆☆  15%
写代码跑实验        ★★☆☆☆  20%（交给AI）
实验分析            ★★★★☆  15%
写论文              ★★★☆☆  5%（AI辅助）
```

## 十四、DL 还需要特征工程吗 & 如何信任别人的代码

### DL 不用手动设计特征，但需要"数据工程"

| 你仍然要做的 | 为什么 |
|-------------|--------|
| 数据质量把控 | 垃圾进垃圾出，标签错了模型再好也白搭 |
| 数据预处理 | 图片 resize、文本分词、数值归一化 |
| 数据增强 | 数据少时翻转/旋转/裁剪增加多样性 |
| 输入格式设计 | 怎么把你的问题编码成模型能吃的格式（需要巧思） |

### 如何信任别人的模型/代码：三步信任法

```
第一步：跑一遍，结果和论文数字对得上吗？
        → 对上 = 基本可信
第二步：读关键代码（模型定义 + 训练循环），和论文描述一致吗？
第三步：在你自己的数据上跑，效果合理吗？
```

**信任指标：** 顶会论文 > 普通期刊 | 官方代码 > 第三方复现 | 高引用 > 低引用 | GitHub 高星 > 低星

### 需要自己造轮子吗？

```
不需要造的：PyTorch、ResNet、BERT、Adam优化器、评估指标库
需要自己写的：你的创新模块、你的数据处理流程、你的实验对比脚本
```

**类比：盖房子不用自己炼钢造砖（用现成材料），但设计图纸是你画的（你的创新）。**

# 深入理解篇：ML 核心机制详解

## 十五、学习率深入解析

### 学习率在更新公式中的角色

```
w_new = w_old - α × 梯度
```

梯度告诉你**方向**（往左还是往右调），学习率 α 决定**你敢走多远**。

同样的梯度 = -100，不同学习率效果完全不同：

```
α = 0.001 → w 变化 = 0.1    （挪了一小步）
α = 0.01  → w 变化 = 1.0    （走了一步）
α = 0.1   → w 变化 = 10.0   （跨了一大步）
α = 1.0   → w 变化 = 100.0  （飞出去了）
```

### 四种学习率的实际效果

假设最优 w = 2.0，当前 w = 0.5：

| 学习率 | 过程 | 结果 |
|--------|------|------|
| 太小 α=0.001 | 每步挪一点点，500 步还没到 | 能收敛但极慢，浪费算力 |
| 合适 α=0.01 | 稳步逼近，约 20 步收敛 | 理想状态 |
| 太大 α=0.1 | 跨过最优点来回震荡，越震越大 | loss 不降反升，剧烈跳动 |
| 爆炸 α=1.0 | 第一步直接飞到 50.5，loss 迅速变 NaN | 训练崩溃 |

### 学习率的本质认知

**学习率本身不提升模型能力。** 模型能力上限由模型结构和数据质量决定。

```
模型结构 + 数据 → 决定了山谷最低点在哪（能力上限）
学习率         → 决定了你能不能走到那个最低点（优化效率）
```

- 学习率合适 → 顺利走到最低点 → 模型发挥全部能力
- 学习率太大 → 跳过最低点 → 有能力但发挥不出来
- 学习率太小 → 卡在半路 → 还没到最低点训练就结束了

**调学习率不是在提升天花板，而是在确保模型能触到自己的天花板。**

### 常见起步值

| 场景 | 常用学习率 |
|------|----------|
| 从零训练小模型 | 0.01 ~ 0.001 |
| 从零训练大模型 | 0.001 ~ 0.0001 |
| 微调预训练模型 | 0.00001 ~ 0.00003 |

微调要小得多，因为预训练参数已经在好的位置，步子大了会破坏已学到的知识。

### 学习率调度（Learning Rate Schedule）

实际训练不用固定学习率，而是动态调整：

```
训练初期：学习率大一点，快速找到大方向
训练中期：逐渐减小，精细调整
训练后期：非常小，微调收敛
```

常见策略：
- **Warmup**：开始用很小的 lr 慢慢增大到目标值，避免初期梯度不稳直接爆炸
- **Cosine Decay**：lr 按余弦曲线从大到小平滑下降
- **Step Decay**：每训练 N 个 epoch，lr 除以 10

### Adam 优化器

Adam 的核心改进：**给每个参数维护自己的"有效学习率"。**
- 梯度一直很大的参数 → 自动调小学习率
- 梯度一直很小的参数 → 自动调大学习率

所以现在大多数人用 Adam，不太需要手动精细调学习率。但初始学习率还是要选对。

## 十六、模型能力提升的四个维度

不同的问题需要调不同的东西：

| 你想做的事 | 该调什么 | 具体手段 |
|-----------|---------|---------|
| 让训练顺利收敛 | 优化过程 | 学习率、优化器（Adam）、batch size |
| 让模型能力上限更高 | 模型结构 | 加层数、加参数量、换更强架构（如 CNN→Transformer） |
| 让模型学到更好的规律 | 数据 | 更多数据、更高质量标注、更好的特征工程 |
| 让模型在新数据上也准 | 泛化能力 | 正则化（L1/L2）、Dropout、数据增强、早停 |

**常见误区**：训练效果不好就一直调学习率。实际上问题可能在数据或模型结构上。

## 十七、训练机制详解

### Epoch、Batch、Iteration

```
数据集有 10000 条
每次取 32 条 → 这 32 条叫一个 Batch（批次）
一个 Epoch = 10000 / 32 = 312 次 Iteration（迭代）
训练 100 个 Epoch = 31200 次参数更新
```

- **Epoch**：所有数据完整过一遍 = 1 个 epoch，通常需要几十到几百个 epoch
- **Batch**：一次送进模型的一批数据，每个 batch 更新一次参数
- **Iteration**：处理一个 batch = 1 次迭代

**不是看完所有数据才更新一次，而是每个 batch 就更新一次。** 这就是为什么叫"随机梯度下降（SGD）"——每次只用一小批数据估计梯度。

### 为什么要多轮（多个 Epoch）

一遍不够。就像看书看一遍记不住，反复看才能吸收。每轮都在已有参数基础上继续优化。

### 从头训练 vs 继续训练 vs 微调

**从头训练（Train from Scratch）**

所有参数从随机数开始，从零学起：

```
w = 0.37（随机生成）→ 经过训练 → w = 2.01 → 保存到 model_v1.pt
```

**继续训练（Resume Training）**

加载之前保存的参数，接着训练：

```
加载 model_v1.pt → w = 2.01 → 继续训练 → w = 2.05 → 保存 model_v2.pt
```

场景：训练中断了不用从头来、数据更新了在旧模型上接着练。

**微调（Fine-tune）**

加载别人训练好的大模型参数，在自己的小数据上继续训练：

```
OpenAI 用万亿 token 训练了 GPT（花了几千万美元）
你加载 GPT 参数 → 在 1000 条医学数据上继续训练几轮
→ 得到一个懂医学的 GPT（只花几小时 + 一张显卡）
```

**现在几乎没人从零训练大模型，全都是微调。**

| | 从头训练 | 继续训练 | 微调 |
|--|---------|---------|------|
| 参数起点 | 随机数 | 自己上次的参数 | 别人训练好的参数 |
| 数据 | 全量数据 | 新增数据或全量 | 自己的少量数据 |
| 成本 | 最高 | 中等 | 最低 |
| 类比 | 新生入学从零学 | 转学生接着学 | 博士在已有知识上学细分领域 |

### Checkpoint（检查点）

训练过程中定期保存参数到文件，防止中断丢失：

```
Epoch 10 → 保存 checkpoint_10.pt
Epoch 20 → 保存 checkpoint_20.pt
...
训练在 Epoch 35 崩了 → 加载 checkpoint_30.pt，从第 30 轮接着来
```

## 十八、训练的随机性与可复现性

### 同样的数据，每次训练结果一样吗？

**不一样。** 三个原因：

1. **参数初始化随机**：不同起点 → 不同收敛路径 → 略不同的最终参数
2. **数据顺序随机（Shuffle）**：每个 epoch 打乱顺序 → 不同 batch 组合 → 不同梯度更新
3. **Dropout 随机**：每次随机关闭不同神经元

### 如何保证可复现？

固定**随机种子（Random Seed）**：

```python
import torch, numpy, random
torch.manual_seed(42)
numpy.random.seed(42)
random.seed(42)
```

种子固定后，所有随机操作结果确定，跑多少次都一样。

**论文和正式实验必须报告 seed 或跑多次取平均，否则结果不可信。**

## 十九、完整实验/科研流程

### 六阶段流程

```
阶段 1：定义问题
  ├── 预测什么？回归还是分类？
  ├── 输入是什么？输出是什么？
  └── 有没有现成数据集？
        ↓
阶段 2：准备数据
  ├── 收集 / 清洗 / 标注
  └── 划分为训练集(70-80%) / 验证集(10-15%) / 测试集(10-15%)
        ↓
阶段 3：选模型 + 设计实验
  ├── 选架构（线性？CNN？Transformer？）
  ├── 确定超参数范围
  └── 先跑一个 Baseline（最简单的方法，作为对照）
        ↓
阶段 4：训练 + 调优（循环）
  ├── 在训练集上训练
  ├── 在验证集上评估
  ├── 好 → 保存模型，试更多改进
  ├── 差 → 分析原因，调整超参或模型
  └── ⚠️ 这个阶段绝不能碰测试集！
        ↓
阶段 5：最终评估
  ├── 模型定型后，拿测试集跑一次（只一次）
  ├── 报告指标（准确率/F1/MSE 等）
  └── 和 Baseline、别人的方法对比
        ↓
阶段 6：分析 + 写论文
  ├── 消融实验（去掉某模块看影响）
  ├── 错误分析（模型在哪类样本上最容易错）
  └── 可视化（损失曲线、混淆矩阵、注意力热图）
```

### 三个数据集的核心区别

```
训练集 → 学生做的练习题（用来学习）
验证集 → 模拟考试（可以考很多次，根据结果调整学习方法）
测试集 → 高考（只考一次，作为最终成绩）
```

验证集不能当测试集：你根据验证集调了模型，模型间接"见过"验证集了，成绩虚高。
测试集是模型从未以任何形式接触过的数据。

**在测试集上调参 = 学术不端级别的错误。**

## 二十、混淆矩阵与评估指标详解

### 为什么需要混淆矩阵

准确率不够用。例：癌症检测，100 人中 95 健康 5 有癌。模型全判"健康"，准确率 95%，但 5 个患者全漏了。

**准确率只告诉你对了多少，不告诉你错在哪。**

### 混淆矩阵结构（以垃圾邮件为例，100 封）

```
                    实际是垃圾邮件    实际是正常邮件
                  ┌──────────────┬──────────────┐
预测是垃圾邮件     │   40 (TP)     │    5 (FP)    │
                  ├──────────────┼──────────────┤
预测是正常邮件     │   10 (FN)     │   45 (TN)    │
                  └──────────────┴──────────────┘
```

| 缩写 | 全称 | 含义 | 本例 |
|------|------|------|------|
| **TP** | True Positive | 预测阳性，实际也阳性——**抓对了** | 40 封垃圾被正确拦截 |
| **FP** | False Positive | 预测阳性，实际阴性——**误杀** | 5 封正常邮件被当垃圾 |
| **FN** | False Negative | 预测阴性，实际阳性——**漏了** | 10 封垃圾没拦住 |
| **TN** | True Negative | 预测阴性，实际也阴性——**正确放行** | 45 封正常邮件正确放行 |

记法：**第一个字母 T/F = 对/错，第二个字母 P/N = 模型的预测。**

### 从混淆矩阵推出所有指标

**准确率** = (TP+TN)/总数 = (40+45)/100 = **85%**

**精确率（Precision）** = TP/(TP+FP) = 40/45 = **88.9%**
→ 模型报警了，可信吗？

**召回率（Recall）** = TP/(TP+FN) = 40/50 = **80%**
→ 该抓的，抓全了吗？

**F1-Score** = 2×(P×R)/(P+R) = 2×(0.889×0.8)/(0.889+0.8) = **84.2%**
→ 精确率和召回率的平衡

### 精确率 vs 召回率：跷跷板关系

| 场景 | 更在乎什么 | 为什么 |
|------|-----------|--------|
| 癌症筛查 | **召回率** | 宁可误诊多检查，不能漏掉患者 |
| 垃圾邮件 | **精确率** | 宁可漏几封垃圾，不能把重要邮件扔垃圾箱 |
| 法律判决 | **精确率** | 宁可放过坏人，不能冤枉好人 |
| 地震预警 | **召回率** | 宁可假警报，不能漏掉真地震 |

**没有"最好"的指标，只有"最适合场景"的指标。**

### 多分类混淆矩阵

分三类（猫/狗/鸟）→ 3×3 矩阵：

```
              实际：猫   实际：狗   实际：鸟
            ┌────────┬────────┬────────┐
预测：猫     │   28   │    3   │    1   │
            ├────────┼────────┼────────┤
预测：狗     │    2   │   25   │    4   │
            ├────────┼────────┼────────┤
预测：鸟     │    0   │    2   │   35   │
            └────────┴────────┴────────┘
```

- 对角线 = 预测正确（28+25+35=88）
- 其他位置 = 模型搞混了的。如：狗被误判为鸟 4 次，猫几乎不会被误判为鸟（0 次）
- **"混淆"矩阵的名字来源：展示模型在哪些类别间发生了混淆**

### 混淆矩阵的诊断价值

1. 哪些类别容易搞混 → 这两类本身很像？需要更多数据或更好特征
2. FN 太高 → 模型太保守，可调低判定阈值
3. FP 太高 → 模型太激进，可调高判定阈值
4. 某一类特别差 → 这一类训练数据太少（类别不平衡问题）

## 二十一、参数与模型文件的本质

### 训练的全部产物就是参数

### 参数的"壳"vs 参数的"值"

- **壳**：在训练前就存在，由模型结构决定（有几层、每层多少节点→决定有多少个参数）

**类比：钢琴出厂时弦的数量就定了（壳），但每根弦的松紧（值）需要调音师来调。**

## 二十二、模型的"记忆"、遗忘与泛化

### 模型如何"记住"知识

模型不会在某处存一句"法国首都是巴黎"。训练时这条数据让某些参数发生微调：

```
看到数据前：w₅₈₇ = 0.421
看到数据后：w₅₈₇ = 0.423（变了 0.002）
```

这个微小变化，与其他几百万个参数的变化**组合在一起**，使模型被问到"法国首都"时能算出"巴黎"概率最高。

**知识分散编码在所有参数里，不存于某个固定位置。**

| 方面 | 人类 | 模型 |
|------|------|------|
| 存储方式 | 神经元连接强度变化 | 参数数值变化 |
| 能精确回忆原文 | 通常不能 | 有时能（其实是过拟合） |
| 能举一反三 | 能 | 能（叫**泛化**） |
| 会遗忘 | 会 | 也会（叫**灾难性遗忘**） |
| 学新东西影响旧知识 | 一般不会 | **会，这是大问题** |

### 灾难性遗忘（Catastrophic Forgetting）

继续训练时，如果新旧数据差别很大，新知识会覆盖旧知识：

```
第一轮：用英语数据训练 → 模型学会了英语
继续：  只用中文数据训练 → 学会了中文，但英语能力急剧下降
```

原因：新数据把参数往新方向调，旧知识对应的参数值被冲掉了。

### 缓解遗忘的方法

| 方法 | 做法 | 原理 |
|------|------|------|
| 混合训练 | 新旧数据混在一起训练 | 两边都有梯度信号，都不丢 |
| 小学习率微调 | 继续训练时用很小的 α | 参数只微调不大改，旧知识保留大部分 |
| 冻结部分参数 | 锁住底层不让动，只训练顶层 | 底层是通用特征（不动），顶层是具体任务（可调） |
| LoRA | 不改原参数，额外加一小组可训练参数 | 原始知识完全不变，新知识存在附加参数里 |

### 泛化（Generalization）

**泛化 = 模型能正确处理它从未见过的新数据。**

这是 ML 的终极目标。模型不是为了在训练数据上考 100 分（那叫死记硬背/过拟合），而是要在**从未见过的新数据**上表现好。

```
训练数据：见过 1000 张猫的照片
泛化能力：第 1001 张从未见过的猫照片，也能正确识别
```

**泛化好 = 学到了真正的规律**（"猫有尖耳朵、竖瞳孔、胡须"）
**泛化差 = 只记住了表面**（"第 37 张图左上角有个水印所以是猫"）→ 过拟合

### 影响泛化的因素

| 因素 | 泛化好 | 泛化差 |
|------|--------|--------|
| 数据量 | 数据多样、覆盖面广 | 数据少、样本单一 |
| 模型复杂度 | 与问题匹配 | 模型太复杂（过拟合）或太简单（欠拟合） |
| 正则化 | 有适当约束（Dropout/L2/早停） | 无约束，模型过度拟合训练数据 |
| 数据增强 | 翻转/旋转/裁剪增加多样性 | 不做增强，模型只见过固定模式 |

### 泛化 vs 过拟合 vs 欠拟合

```
欠拟合：训练集上就表现差 → 没学会 → 模型太简单或训练不够
过拟合：训练集上好，新数据上差 → 死记硬背 → 模型太复杂或数据太少
泛化好：训练集和新数据上都好 → 学到了真正的规律 → 这是目标
```

**一句话：训练是手段，泛化才是目的。模型的价值不在于它在训练数据上多准，而在于面对从未见过的数据时依然准确。**
